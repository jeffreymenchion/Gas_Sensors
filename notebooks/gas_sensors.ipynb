```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame()

df = pd.read_csv("/Users/jeff/Documents/datasets/air_quality/air+quality/AirQualityUCI_spreadsheet1.csv")
df
```

```python
#Formatting some object columns from strings to floats

df.replace(to_replace=',',value='.',regex=True,inplace=True)

for i in 'CO(GT) C6H6(GT) T RH AH'.split():
    df[i] = pd.to_numeric(df[i], errors='coerce')

df
```

```python

#Formatting Date and Time to datetime type
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

df['date'] = pd.to_datetime(df['Date'])
df['Day_of_Week'] = df['date'].dt.day_name()
df
```

```python
#Altering the Dataframe - Aligning the Day_of_Week with Date

df = df[['Date', 'Day_of_Week', 'Time', 'CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']]
df
```
